## Data Vizualization

https://www.data-to-viz.com/


#### Univariate analysis

```python
plt.figure(figsize=(7, 5))

# Plot KDE
sns.kdeplot(data=df, x="bmi", fill=True, color="crimson")

# Customize labels
plt.title("KDE Curve: BMI Distribution")
plt.xlabel("Body Mass Index (Scaled)")
plt.ylabel("Density")

plt.tight_layout()
plt.show()
```

#### Bivariate analysis

```python
plt.figure(figsize=(7, 5))

# Plot regression
sns.regplot(
    data=df,
    x="bmi",
    y="target",
    scatter_kws={"alpha": 0.5},
    line_kws={"color": "red"},
)

# Customize labels
plt.title("Regression Curve: BMI vs Progression Target")
plt.xlabel("Body Mass Index (Scaled)")
plt.ylabel("Disease Progression")

plt.tight_layout()
plt.show()
```

#### Multivariate analysis

```python
plt.figure(figsize=(10, 6))

# Plot boxplot for all features except target
sns.boxplot(data=df.drop(columns=["target"]), orient="v", palette="Set2")

# Customize labels
plt.title("Box Plot of Diabetes Dataset Features", fontsize=14)
plt.xlabel("Features", fontsize=12)
plt.ylabel("Value", fontsize=12)
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```
